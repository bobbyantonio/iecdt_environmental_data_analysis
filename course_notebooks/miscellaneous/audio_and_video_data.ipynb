```python
import os
from tqdm import tqdm
import numpy as np
from moviepy.editor import VideoFileClip
import matplotlib.pyplot as plt
from pathlib import Path

notebook_dir = os.getcwd()
data_dir = str(Path(notebook_dir).parents[1] / 'data_samples' )
```

```python
video_clip = VideoFileClip(os.path.join(data_dir, 'audio_visual', 'video_with_sound_example.mp4'))
audio_clip = video_clip.audio
```

```python
# Play the video clip
video_clip.ipython_display(width=500, maxduration=100)
```

```python
# Iterate through the video clips
# As an example, lets calculate the average intensity at each frame
averages = []
for frame in tqdm(audio_clip.iter_frames()):
    
    averages.append(frame.max())
plt.plot(averages)
```

```python
# Length of video clip
video_clip.duration
```

```python
# Frames per second
video_clip.fps
```

```python
# Number of frames
video_clip.reader.nframes
```

```python
# Selecting frames by time (in seconds)
frame = video_clip.get_frame(30)

# Extract frame by index; moveipy doesn't seem to support this natively, so you have to use the frames-per-second property
frame = video_clip.get_frame(100 / video_clip.fps)
```
