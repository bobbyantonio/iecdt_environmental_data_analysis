```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Level one
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/simple_plot.html#sphx-glr-gallery-lines-bars-and-markers-simple-plot-py
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Create the axes
ax = plt.axes()

# Plot the data
ax.plot(t, s)

# Customize the plot
ax.set_xlabel("time (s)")
ax.set_ylabel("voltage (mV)")
ax.set_title("Level 1")
ax.grid()

# Save and show
plt.savefig("examples/level1.png")
plt.show()
```

```python
# Level 2
# Data for plotting
t = np.arange(0.0, 10.0, 0.01)
s = np.sin(2 * t)

# Create the axes
ax = plt.axes()

# Plot the data
ax.plot(t, s, linestyle = '-.', color = "k",)

# Customize the plot
ax.set_xlabel("time (s)")
ax.set_ylabel("voltage (mV)")
ax.set_title("Level 2", fontsize = 15, color = "red")
ax.set_xticks(np.arange(0,10,np.pi), ['0', '', '2', '3',])
ax.grid()

# Save and show
plt.savefig("examples/level2.png")
plt.show()
```

```python
# Level 3
# https://matplotlib.org/stable/gallery/pyplots/pyplot_text.html#sphx-glr-gallery-pyplots-pyplot-text-py
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

# Create the axes
ax = plt.axes()

# Plot the data
ax.plot(t, s)

# Add some text
plt.text(0, -1, r'Hello, world!', fontsize=15)

# Add latex in the title
plt.title(r'Level 3: $\sin(\omega t)$', fontsize=20)

# Axes labels
plt.xlabel('Time [s]')
plt.ylabel('Voltage [mV]')

# Save and show
plt.savefig("examples/level3.png")
plt.show()
```

```python
# Data for plotting
t = np.arange(0.01, 10.0, 0.01)
e = np.exp(t)
s = np.sin(2 * np.pi * t)

# Create the axes
ax1 = plt.axes()

# Plot the exponential on ax1 in red
ax1.plot(t, e, color="red")

# Twin ax1 into ax2
ax2 = ax1.twinx()  

# Plot the sinus on ax2 in blue
ax2.plot(t, s, color="blue")

# Label the axes
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color="red")
ax2.set_ylabel('sin', color="blue")  

# Color the ticks
ax1.tick_params(axis='y', labelcolor="red")
ax2.tick_params(axis='y', labelcolor="blue")

# Set the title
plt.title("Level 4: Double y-axis")

# Save and show
plt.savefig("examples/level4.png", bbox_inches = "tight")
plt.show()
```

```python
#import matplotlib.collections as mcol
#from matplotlib.legend_handler import HandlerLineCollection, HandlerTuple
#from matplotlib.lines import Line2D

# Data for plotting
t = np.arange(0.0, 2.0, 0.05)
y1 = np.exp(-t)
y2 = np.sin(2 * np.pi * t)
y3 = np.exp(-t) * np.sin(2 * np.pi * t)

# Create axes
ax = plt.axes()

# Plot the data
l1 = ax.plot(t, y1, color = 'gray', linestyle = ':')
l2 = ax.plot(t, y2, color = 'k', linestyle = '--', marker = 'o')
l3 = ax.plot(t, y3, linestyle = '-.', marker = '+')

# Add the legend
ax.legend(("exponential", 'oscillatory', 'damped'), loc='upper right', shadow=True)

# Add labels
ax.set_xlabel('time')
ax.set_ylabel('volts')
ax.set_title('Level 5: Labels')

# Save and show
plt.savefig("examples/level5.png")
plt.show()
```

```python
fig, axs = plt.subplot_mosaic([['a)', 'c)'], ['b)', 'c)'], ['d)', 'd)']],
                              layout='constrained')
for label, ax in axs.items():
    # Use Axes.annotate to put the label
    # - at the top left corner (axes fraction (0, 1)),
    # - offset half-a-fontsize right and half-a-fontsize down
    #   (offset fontsize (+0.5, -0.5)),
    # i.e. just inside the axes.
    ax.annotate(
        label,
        xy=(0, 1), xycoords='axes fraction',
        xytext=(+0.5, -0.5), textcoords='offset fontsize',
        fontsize='medium', verticalalignment='top', fontfamily='serif',
        bbox=dict(facecolor='0.7', edgecolor='none', pad=3.0))
```
