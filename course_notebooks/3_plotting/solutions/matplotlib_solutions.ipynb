```python
import matplotlib.pyplot as plt
import numpy as np
```

# Matplotlib

```python
# Level one
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/simple_plot.html#sphx-glr-gallery-lines-bars-and-markers-simple-plot-py
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Create the axes
ax = plt.axes()

# Plot the data
ax.plot(t, s)

# Customize the plot
ax.set_xlabel("time (s)")
ax.set_ylabel("voltage (mV)")
ax.set_title("Level 1")
ax.grid()

# Save and show
plt.savefig("examples/level1.png")
plt.show()
```

```python
# Level 2
# Data for plotting
t = np.arange(0.0, 10.0, 0.01)
s = np.sin(2 * t)

# Create the axes
ax = plt.axes()

# Plot the data
ax.plot(t, s, linestyle = '-.', color = "k",)

# Customize the plot
ax.set_xlabel("time (s)")
ax.set_ylabel("voltage (mV)")
ax.set_title("Level 2", fontsize = 15, color = "red")
ax.set_xticks(np.arange(0,10,np.pi), ['0', 'π', '2π', '3π',])
ax.grid()

# Save and show
plt.savefig("examples/level2.png")
plt.show()
```

```python
# Level 3
# https://matplotlib.org/stable/gallery/pyplots/pyplot_text.html#sphx-glr-gallery-pyplots-pyplot-text-py
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

# Create the axes
ax = plt.axes()

# Plot the data
ax.plot(t, s)

# Add some text
plt.text(0, -1, r'Hello, world!', fontsize=15)

# Add latex in the title
plt.title(r'Level 3: $\sin(\omega t)$', fontsize=20)

# Axes labels
plt.xlabel('Time [s]')
plt.ylabel('Voltage [mV]')

# Save and show
plt.savefig("examples/level3.png")
plt.show()
```

```python
# Data for plotting
t = np.arange(0.01, 10.0, 0.01)
e = np.exp(t)
s = np.sin(2 * np.pi * t)

# Create the axes
ax1 = plt.axes()

# Plot the exponential on ax1 in red
ax1.plot(t, e, color="red")

# Twin ax1 into ax2
ax2 = ax1.twinx()  

# Plot the sinus on ax2 in blue
ax2.plot(t, s, color="blue")

# Label the axes
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color="red")
ax2.set_ylabel('sin', color="blue")  

# Color the ticks
ax1.tick_params(axis='y', labelcolor="red")
ax2.tick_params(axis='y', labelcolor="blue")

# Set the title
plt.title("Level 4: Double y-axis")

# Save and show
plt.savefig("examples/level4.png", bbox_inches = "tight")
plt.show()
```

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.05)
y1 = np.exp(-t)
y2 = np.sin(2 * np.pi * t)
y3 = np.exp(-t) * np.sin(2 * np.pi * t)

# Create axes
ax = plt.axes()

# Plot the data
l1 = ax.plot(t, y1, color = 'gray', linestyle = ':')
l2 = ax.plot(t, y2, color = 'k', linestyle = '--', marker = 'o')
l3 = ax.plot(t, y3, linestyle = '-.', marker = '+')

# Add the legend
ax.legend(("exponential", 'oscillatory', 'damped'), loc='upper right', shadow=True)

# Add labels
ax.set_xlabel('time')
ax.set_ylabel('volts')
ax.set_title('Level 5: Labels')

# Save and show
plt.savefig("examples/level5.png")
plt.show()
```

```python
# Inspired from: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html#sphx-glr-gallery-subplots-axes-and-figures-subplots-demo-py

# Data for plotting
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# Create figure with two rows of three subplots, sharing x-axis, in a figure of size [10,4]
fig, axs = plt.subplots(2,3, sharex = True, figsize = [10,4])

# Plot the data on the axes
axs[0,0].plot(x,y, color = 'k', linewidth = 1)
axs[1,0].plot(x,-y, color = 'k', linewidth = 1)
axs[0,1].plot(x,y/2, color = "grey", linewidth = 0.5)
axs[1,1].plot(x,-y*2, color = "k", linewidth = 2)
axs[0,2].plot(x,y+1, color = "tab:red", linewidth = 1)
axs[1,2].plot(x+1,y, color = "tab:green", linestyle = '--')

# Subplot titles
axs[0,0].set_title("y vs. x")
axs[1,0].set_title("-y vs. x")
axs[0,1].set_title("y/2 vs. x")
axs[1,1].set_title("-y*2 vs. x")
axs[0,2].set_title("y+1 vs. x")
axs[1,2].set_title("y vs. x+1")
# Figure title
fig.suptitle("Level 6: Subplots")

# Save and show
plt.tight_layout()
plt.savefig("examples/level6.png")
plt.show()
```

```python
# Data for plotting
t = np.arange(0.0, 10.0, 0.05)
e = np.exp(t)

# Create figure with two subplots
fig, axs = plt.subplots(1, 2, figsize = [6,3]) # Two columns of one subplot

# Plot the data on both axes
axs[0].plot(t,e)
axs[1].plot(t,e)

# Set the y-axis of the second subplot as log
axs[1].set_yscale("log")

# Add labels to both subplots
for i in range(len(axs)):
    axs[i].set_xlabel('x')
    axs[i].set_ylabel('y')

# Subplot titles
axs[0].set_title("Linear y-axis")
axs[1].set_title("Logarithmic y-axis")
# Figure title
fig.suptitle('Level 7: Log axes')

# Save and show
plt.tight_layout()
plt.savefig("examples/level7.png")
plt.show()
```

```python
# Inspired from: https://stackoverflow.com/questions/31345489/change-color-of-lineplot-depending-on-data

# Data for plotting
x = np.linspace(0, 30, 2000)
signal = 1.2*np.sin(x)

# Create axes
ax = plt.axes()

# Mask
pos_signal = np.where(signal>=0, signal, np.nan)
neg_signal = np.where(signal<=0, signal, np.nan)

# Plotting with linewidth=3
ax.plot(pos_signal, color='r', linewidth  = 3)
ax.plot(neg_signal, color='b', linewidth  = 3)

# Customize
ax.set_title("Level 8: Color line by sign")

# Save and show
plt.savefig('examples/level8.png',)
plt.show()
```

```python
# As an example, we will extract again the time series of temperature over Oxford
import xarray as xr
ds = xr.open_dataset("../../data_samples/netcdf/E-OBS/UK_monthly.nc") # Open the dataset
T = ds.tg # Select the temperature variable
T_Oxford_monthly = T.interp(latitude = 51.75, longitude = -1.26,) # Interpolate the data for Oxford
T_Oxford_yearly = T_Oxford_monthly.groupby("time.year").mean()
```

```python
# Data for plotting
T_Jan = T_Oxford_monthly.where(T_Oxford_monthly.time.dt.month == 1, drop = True) # Select the temperature for the months of January only

# Create axes
ax = plt.axes()

# Plot the temperature with coloring
ax.bar(T_Jan.time.dt.year, T_Jan, color=np.where(T_Jan > 0, 'b', 'r'),)

# Labels 
ax.set_xlabel("Year")
ax.set_ylabel("Temperature / °C")
ax.set_title("Level 9: Colored bar plot on Jan. T in Oxford")

# Save and show
plt.savefig('examples/level9.png',)
plt.show()
```

```python
import matplotlib.ticker as ticker

# Data for plotting
p95 = np.percentile(T_Oxford_monthly, 98)
p5 = np.percentile(T_Oxford_monthly, 2)

# Create figure of size [15,4] and axes
plt.figure(figsize = [15,4])
ax = plt.axes()

# Plot Oxford monthly temperature data
ax.plot(T_Oxford_monthly.time, T_Oxford_monthly, color = "k", zorder= 1)

# Add points for extreme temperatures
ax.scatter(T_Oxford_monthly.time, T_Oxford_monthly.where(T_Oxford_monthly < p5), color = "tab:blue", zorder = 10, label = "2% hottest months", marker = '*')
ax.scatter(T_Oxford_monthly.time, T_Oxford_monthly.where(T_Oxford_monthly > p95), color = "tab:red", zorder = 10, label = "2% coldest months", marker = '*')

# Labels axes
ax.set_xlabel("Time")
ax.set_ylabel("Temperature (°C)")

# Add info about records
ax.annotate('', xy=(T_Oxford_monthly.idxmax(),-6), xytext=(T_Oxford_monthly.idxmax(),T_Oxford_monthly.max()-0.5),
            arrowprops=dict(facecolor='tab:red', alpha = 0.8, width = 2), zorder = 20, 
            )
ax.annotate('', xy=(T_Oxford_monthly.idxmin(),-6), xytext=(T_Oxford_monthly.idxmin(),T_Oxford_monthly.min()-0.5),
            arrowprops=dict(facecolor='tab:blue', alpha = 0.8, width = 2, headlength = 5,), zorder = 20, 
            )
ax.axhline(y= T_Oxford_monthly.max(), color = 'tab:red')
ax.axhline(y= T_Oxford_monthly.min(), color = 'tab:blue')
ax.text(x = T_Oxford_monthly.time.min(), y = T_Oxford_monthly.max() + .5, s = "Highest monthly Temperature: "+ str(np.round(T_Oxford_monthly.max().values,1)) + " in " + str(T_Oxford_monthly.idxmax().values)[:7], fontweight="bold", color = "tab:red")
ax.text(x = T_Oxford_monthly.time[170], y = T_Oxford_monthly.min() + .5, s = "Lowest monthly Temperature: "+ str(np.round(T_Oxford_monthly.min().values,1)) + " in " + str(T_Oxford_monthly.idxmin().values)[:7], fontweight="bold", color = "tab:blue")

# Setup axes
ax.set_xticks(T_Oxford_monthly.time[::12], minor = True)
ax.set_ylim(-6,24)

# Legend
ax.legend(loc = "lower right")

# Title
ax.set_title("Level 10: Illustrating Oxford's T time series (difficult)")

# Save and show
plt.tight_layout()
plt.savefig('examples/level10.png', dpi = 1000)
plt.show()
```

```python
import cartopy.crs as ccrs
```

```python
# Data for plotting 
delta = 0.2
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

# Create figure with three subplots (figsize = [14,4])
fig, axs = plt.subplots(1, 3, figsize = [14,4])

# In the first one, plot the data with pcolormesh
p = axs[0].pcolormesh(X,Y,Z)
plt.colorbar(p)
axs[0].set_title("pcolormesh")

# In the second one, plot the data with contourf
p = axs[1].contourf(X,Y,Z)
plt.colorbar(p)
axs[1].set_title("contourf")

# In the third one, plot the data with contour
p = axs[2].contour(X,Y,Z)
plt.colorbar(p)
axs[2].set_title("contour")

# Title
fig.suptitle("Level 1: Compare three different way to plot 2D data")

# Save and show
plt.tight_layout()
plt.savefig('examples/maps_level1.png',)
plt.show()
```

```python
# Inspired from: https://matplotlib.org/stable/gallery/images_contours_and_fields/contour_demo.html#sphx-glr-gallery-images-contours-and-fields-contour-demo-py

# Data for plotting 
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

# Create axes
ax = plt.axes()

# Plot contour with labels
c = ax.contour(X, Y, Z, colors = 'k')
ax.clabel(c, inline=True, fontsize=10)

# Title
ax.set_title("Level 2: Contour with labels")

# Save and show
plt.tight_layout()
plt.savefig('examples/maps_level2.png',)
plt.show()
```

```python
# Data for plotting 
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

# Create axes
ax = plt.axes()

# Plot contour with labels
p = ax.contourf(X, Y, Z, cmap= "PiYG", levels = np.arange(-2,2.1,0.25))
plt.colorbar(p)

# Title
ax.set_title("Level 3: contourf with specific levels")

# Save and show
plt.tight_layout()
plt.savefig('examples/maps_level3.png',)
plt.show()
```

```python
from matplotlib.colors import BoundaryNorm

# Data for plotting 
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

# Create axes
ax = plt.axes()

# Plot contour with labels
p = ax.pcolormesh(X, Y, Z, cmap= "PiYG", norm = BoundaryNorm(np.arange(-2,2.1,0.5,), 256))
plt.colorbar(p)

# Title
ax.set_title("Level 4: pcolormesh with specific levels")

# Save and show
plt.tight_layout()
plt.savefig('examples/maps_level4.png',)
plt.show()
```

```python
# From https://scitools.org.uk/cartopy/docs/v0.16/gallery/waves.html#sphx-glr-gallery-waves-py

# Data for plotting
nlats, nlons = (73,145)
lats = np.linspace(-np.pi / 2, np.pi / 2, nlats)
lons = np.linspace(0, 2 * np.pi, nlons)
lons, lats = np.meshgrid(lons, lats)
wave = 0.75 * (np.sin(2 * lats) ** 8) * np.cos(4 * lons)
mean = 0.5 * np.cos(2 * lats) * ((np.sin(2 * lats)) ** 2 + 2)

lats = np.rad2deg(lats)
lons = np.rad2deg(lons)
data = wave + mean

# Create the axes
ax = plt.axes(projection = ccrs.Mollweide())

# Plot the data
p = ax.contourf(lons, lats, data,
            transform=ccrs.PlateCarree(),
            cmap='Spectral')

# Add colorbar
plt.colorbar(p, shrink = 0.5)

# Add coastlines 
ax.coastlines()

# Add title 
ax.set_title("Level 5: Wavy map")

# Save and show
plt.tight_layout()
plt.savefig('examples/maps_level5.png',)
plt.show()
```
